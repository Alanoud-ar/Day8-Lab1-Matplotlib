## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,0,4,5],color = '#EE82EE');
plt.plot([10,9,8,6,3],color = '#7E1E9C');
plt.plot([5,7,9,3,1],color = '#ED0EE9');
# Add Legends
plt.legend(['2008','2009','2010'], loc = 'upper right') # location (0 = best);
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(1,3,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

#Third
plt.subplot(1,3,3)
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x,y,alpha = 0.5,s = 50);
plt.title('Scatter Chart', loc='left', color='r');
plt.xlabel('Days', fontsize = 16);
plt.ylabel('Sales', fontsize = 16);
```
